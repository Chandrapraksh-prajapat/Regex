```python
# Q4. What is hstack and vstack , explain with code of example?


# hstack => hstack is a numpy in-built function which is used to add arrays horizontally but it should have same number of rows.
# example:

a=np.array([1,2,3])
b=np.array([4,5,6])
np.hstack((a,b))
```

```python
# Q5.What is view vs copy shoe me with the example

# view=> a view share memory with an array it means any change in view will reflect in original array.

# example :

a=np.array([1,2,3])
b=a.view()
b[0]=10
a
```

```python
# copry=>  Copy means create a copry of original array ,
#         a copy does not share memory with an array it means any change in copy will not reflect in original array.

# example :

a=np.array([1,2,3])
b=a.copy()
b[0]=10
a
```

```python
# Q7. You have too create an array in range of 1 to 10 and then your have to reshape it (3,3) and then perform total sum , rowise sum , column wise sum

a=np.arange(1,10).reshape(3,3)
print(a)

# total sum :
print("Total : ",np.sum(a))

# Row wise sum :
print("Row wise : ", np.sum(a,axis=1))

# Column wise sum :
print("Column wise : ",np.sum(a,axis=0))
```
