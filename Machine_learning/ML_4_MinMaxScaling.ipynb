```python
df=pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head(3)
```

```python
df=df.dropna()
```

```python
df=df.drop(columns=['city'])
```

```python
df.head(2)
```

```python
# step 1 :: convert categorical data into numerical data

df['gender']=df['gender'].map({'Male':0,'Female':1})
df['cough']=df['cough'].map({'Mild':0,'Strong':1,'Strong':2})
df['has_covid']=df['has_covid'].map({'Yes':1,'No':0})
df

```

```python
## step-2 data dividation

x=df.drop(columns=['has_covid'])
y=df['has_covid']

```

```python
x.head(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
np.round(x_train.describe(),2)
```

```python
## step -3 Apply MinMaxScaling
```

```python
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
mn
```

```python
x_train_mn=mn.fit_transform(x_train)
```

```python
x_train_new=pd.DataFrame(x_train_mn, columns = x.columns)
```

```python
np.round(x_train_new.describe(),2)
```
