```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head(3)
```

```python
# df.isnull().sum()
```

```python
df['fever']=df['fever'].fillna(df['fever'].mean())
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['city'])
df
```

```python
df['gender']=df['gender'].map({'Male':1,'Female':0})
df['cough']=df['cough'].map({'Mild':1,'Strong':0})
df['has_covid']=df['has_covid'].map({'Yes':1,'No':0})
df
```

```python
x=df.drop(columns=['has_covid'])
x
```

```python
y=df['has_covid']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
np.round(x_train.describe(),2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)  # fit means learn the parameters and transform means apply on that
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
x_train_new
```
