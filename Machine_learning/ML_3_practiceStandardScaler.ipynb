```python
df=pd.read_csv("/content/tips - tips.csv")
df.head(3)
```

```python
df=df.drop(columns=['day','time'])
df
```

```python
df['sex']=df['sex'].map({'Female':0,'Male':1})
df['smoker']=df['smoker'].map({'No':0,'Yes':1})
df
```
