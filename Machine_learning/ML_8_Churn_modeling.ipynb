```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/Churn_Modelling - Churn_Modelling.csv")
df.head(2)
```

```python
df=df.drop(columns=['RowNumber','CustomerId','Surname','Tenure'])
df.head(2)
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
x=df.drop(columns=['Exited'])
y=df['Exited']

x.head(2)
# y.head(2)
```

```python
# # Encoding (Label , Ordinal , Onehot , get_dummies)

## Label encoding
#  gender --->> Male / Female ----> 0 / 1

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

x['Gender_Label'] = le.fit_transform(x['Gender'])

x.head(2)

```

```python
## Ordinal encoding Geography

from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()

x['Geography_Ordinal'] = oe.fit_transform(x[['Geography']])

x.head(2)
```

```python

## One hot encoding

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)

Geo_encode=ohe.fit_transform(x[['Geography']])



Geo_encode_df=pd.DataFrame(Geo_encode,columns=ohe.get_feature_names_out(['Geography']))

Geo_encode_df.head(2)
```

```python
x=pd.concat([x,Geo_encode_df],axis=1)
x.head(2)
```

```python
# get_dummies

x=pd.get_dummies(x,columns=['Gender'],drop_first=True)
x.head(2)
```

```python
## standardization

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

num_cols=['CreditScore','Age','Balance','NumOfProducts','EstimatedSalary']

x[num_cols]=sc.fit_transform(x[num_cols])

x.head(2)
```

```python
#  normalization

from sklearn.preprocessing import MinMaxScaler

mn=MinMaxScaler()

x[num_cols]=mn.fit_transform(x[num_cols])

x.head(2)
```
