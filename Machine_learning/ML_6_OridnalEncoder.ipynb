```python
# LabelEncoder ---> when we have single column . we cannot set sub-category form .
# has_covid['yes','no'] ---> labelecoder --> [1,0]
# OrdinalEncoder ---> when we have multiple columns .we can set sub-category form.
# has_covid['yes','no'] ---> ord. encoder ---> [0,1]
```

```python
# first approach ::::
# df ---> categorical column ---> ordinal encoder ---> data transform
# second approach ::::
# df ---> x,y --->
# x ---> encode by Ordinal Encoder
# y ---> encode by Label Encoder
```
