```python
# SET : A Set is a built-in Python data type used to store multiple unique values in a single variable.

# Sets automatically remove duplicate values.

# Sets are written using curly braces { }

# Sets are:

#  Unordered
#  Mutable
#  Do NOT allow duplicates
#  Do NOT support indexing


# Why Sets are Used?

#  Remove duplicates
#  Mathematical operations
#  Membership testing (in) is fast

# Creating Set:

myset = {1,2,3}
print(myset)
```

```python
numbers = {1,2,2,3,3,4}  #Duplicate Removal
print(numbers)
```

```python
s = {10,20,30} # Add & Remove

s.add(40)     # it will add at the starting position
s.remove(20)

print(s)
```

```python
# Common Set Methods (Must Know)

# add()
# remove()
# union()
# intersection()
# difference()


a = {1,2,3}
b = {3,4,5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

```python
#  Create set of numbers and remove duplicates.

numbers = [1,2,2,3,3,4]
unique_numbers = set(numbers)
print(unique_numbers)
```

```python
#  Find common elements between two sets.

set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}

print(set1.intersection(set2))
```
