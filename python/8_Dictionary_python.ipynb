```python
# Common Dictionary Methods (Must Know):


# keys()
# values()
# items()
# get()
# pop()


d = {"x":10,"y":20}

print(d.keys())
print(d.values())
print(d.items())
print(d.get("x"))
print(d.pop("y"))
print(d)
```

```python
emp = { "name": "Amit",  "salary": 40000,  "city": "Delhi"}

print(emp["name"])

emp["salary"] = 45000
print(emp)
```

```python

data = {"a":10,"b":20,"c":30}

for k,v in data.items():
    print(k,v)

```

```python
# count the element in range & assign the dictionary

num ={ "count" :0}

for i in range( 1, 11):
  # num["count"] = i
  num['count'] = num["count"]+1
print(num)
```

```python
# count prime values

number ={"prime" : 0}
for num in range(1,8):
  count=0

  for i in range(1,num+1):
    if num%i==0:
      count=count+1

  if count==2:
      number["prime"] = number["prime"]+1
print(number)

```

```python
# print the keys in dictionary
xyz = {"count" : 0 , "run" : 55 }

for i in xyz.keys():
  print(i)
```

```python
# print values in dictionary

xyz = { "name" : "Rahul" , " course " : "DS "}

for i in xyz.values():
  print(i)
```
