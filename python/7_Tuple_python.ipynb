```python
# Tuple : A Tuple is a built-in Python data type used to store multiple values in a single variable,
#            just like a list â€” but tuples are immutable (cannot be changed after creation).



# Tuples are written using round brackets ( )
# Tuples are ordered
# Tuples are immutable (no add / remove / update)
# Tuples allow duplicate values
# Tuples can store different data types


# Tuples are used when:

#  Data should NOT change
#  Faster than list
#  Used for fixed records (like database rows)
```

```python
# Creating a tuple:

my_tuple =(10,20,30)
print(my_tuple)
print(my_tuple[0])
print(type(my_tuple))
```

```python
data = ("Rahul", 21, "Jaipur")

print(data)
print(data[0])
```

```python
numbers = (10, 20, 30)

for n in numbers:
    print(n)
```

```python
# Tuple Methods (Very Few)

# count(): count the same element in the tuple
# index(): return the index position of given element


t = (10,20,10,30)
print(t.count(10))
print(t.index(20))
```

```python
# Why are tuples faster than lists?

# Tuples are immutable, so Python can optimize memory allocation and
#   does not need to handle dynamic resizing or modification operations.
#        This makes tuple operations slightly faster than lists.


```
