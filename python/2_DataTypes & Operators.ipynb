```python
# Python has several built-in data types:

# Numeric:
        # int → whole numbers
        # float → decimal numbers
        # complex → numbers like 3+4j

# Text:
      # str → strings (text)

# Sequence:
          # list → changeable collections
          # tuple → unchangeable collectionss
          # range → sequences of numbers

# Mapping:
          # dict → key-value pairs

# Set:
      # set → unique items
      # frozenset → unchangeable set

# Boolean:
          # bool → True/False

# Binary:
          # bytes, bytearray, memoryview
```

```python
# Operators in python :
                      # Arithmetic
                      # Comparision
                      # Assignment
                      # Logical
                      # Membership
                      # Identity

```

```python
# Arithmetic Operator : use to perform basic mathematical operations
a=10
b=3
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a//b)
print(a**b)
```

```python
# cpmparision

x=5
y=10
print(x==y)
print(x!=y)
print(x>y)
print(x<y)
print(x>=y)
print(x<=y)
```

```python
# Assignment

num=10

num+=5
print(num)
num-=3
print(num)

num*=2
print(num)
num/=4
print(num)

num%=5
print(num)
num//=2
print(num)
```

```python
# Logical operator

a=True
b=False
print(a and b)
print(a or b)
print(not a)
```

```python
# Membership  : a value exists in a sequence (list , string etc.)

nums=[1,2,3,4]

print( 2 in nums)
print(20 in nums)
print(5 not in nums)
```

```python
# Identity operators : two variable point to the same object in memory

a=[1,2,3]
b=a
c=[1,2,3]

print(a is b)
print(a is c)
print(a is not c )
```
