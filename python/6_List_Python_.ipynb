```python
# Find odd number and store in a new list

a=[10,11,12,15,19]
b=[]
for i in range(len(a)):
  if a[i]%2!=0:
   b.append(a[i])
print("odd list :",b)
```

```python
a=[10,11,12,15,19]
b=a[0]
for i in range(0,len(a)):
  curr=a[i]
  if (curr>b):
   b=curr
print("Max value:",b)
```

```python
# Sum of elements in a new list
a=[10,11,12,15,19]
b=[]
sum=0
for i in range(len(a)):
  curr=a[i]
  sum=sum+curr
  b.append(sum)
print(b)
```

```python
# find duplicates in a list and store in a new list
a=[10,11,12,15,19,10,12]
b=[]
duplist=[]
for i in range(len(a)):
  curr=a[i]
  if curr not in b:
    b.append(curr)
  else:
    duplist.append(curr)
print(duplist)
```

```python
# Find vowel and consonent in a string and store in a new list
a='ishan hey'
vowel=[]
cons=[]
for i in range(0,len(a)):
  curr=a[i]
  if curr in 'aeiouAEIOU':
    vowel.append(curr)
  else:
    cons.append(curr)
print(vowel)
print(cons)
```

```python
# Store missing numbers in a new list in a given range
a=[1,2,4,6,7,9]
n=9
missingnum=[]
for i in range(1,n+1):
  if i not in a:
    missingnum.append(i)
print(missingnum)
```

```python
# find minimum and maximum element of list without using function

a=[1,2,3,4,5,6,7,8,9]
min=a[0]
max=a[0]
for i in a:
  if i>max:
    max=i
  if i<min:
    min=i
print(min)
print(max)
```

```python
# find integers from a list and calculate sum
a=['sam',45,'raj',29.67,10]
sum=0
for i in a:
  if type(i)==int:
    sum=sum+i
print(sum)
```
