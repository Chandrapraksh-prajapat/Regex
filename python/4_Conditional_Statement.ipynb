```python
# # In Python, conditional statements are used to make decisions based on conditions

# Types of Conditional Statements

# if
# if-else
# if-elif-else

# Nested if
```

```python
# 1️. if Statement

# Executes code only if condition is True

age = 18

if age >= 18:
    print("You are eligible to vote")
```

```python
# 2️. if-else Statement: if condition is true print true part otherwise print false part
                      # Chooses between two conditions

num = 7
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
# 3️. if-elif-else Statement :  Used when there are multiple conditions

marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

```python
# 4️. Nestedd if : if inside another if

num = 10

if num > 0:
    if num % 2 == 0:
        print("Positive Even")
```

```python
# if (One-line):
a = 10
b = 20

if a < b: print("b is greater")
```

```python
# Ternary Conditional (Short form)

a = 5
b = 10

result = "a is big" if a > b else "b is big"
print(result)
```

```python
# #  Comparison Operators :-

# ==   equal to
# !=   not equal
# >    greater than
# <    less than
# >=   greater or equal
# <=   less or equal
```
