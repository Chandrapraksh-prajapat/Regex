```python
#CSV = Comma Separated Values

# It is a file format used to store tabular data (rows & columns).

# Example csv file:
    # name,age,city
    # Rahul,21,Jaipur
    # Amit,22,Delhi
    # Neha,20,Mumbai

# Each line = row
# Values separated by ,
```

```python
# Why CSV is Important?

# Used in Excel
# Used in Data Analysis
# Used in datasets
# Used before Pandas
```

```python
# Python CSV Module:-

# Python has built-in module used to read & write csv files.

import csv
# No need to install anything.

# read csv files row by row
# write data into csv files
# supports dictionary style read / write
# handles comman quotes & special characters
# works with excel-friendly file
# makes tabuler data processing easy
```

```python
# main features of csv module:

# 1.read csv file :

import csv

with open("file.csv","r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
import csv

with open("students.csv","r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
# 2.writing csv
import csv
with open("file.csv","w",newline="") as file:
  data=csv.writer(file)
  data.writerow(["name","age","city"])
  data.writerow(["Rahul",21,"Jaipur"])
```

```python
# Append csv : Adds new row without deleting old data

# Note : Already store data are clear then write new data

# if you want to not remove old data & insert new data in file & insert to line
# use mode='a'

# Example:

import csv
with open("file1.csv","a",newline="") as file:
  data=csv.writer(file)
  data.writerow(["name","age"])
  data.writerow(["Rahul",21])
  data.writerow(["Amit",22])


# read data from file
with open("file1.csv","r") as file:
  reader = csv.reader(file)

  for row in reader:
    print(row)
```

```python
# DictReader (VERY IMPORTANT)

# Instead of list → dictionary

import csv

with open("students.csv","r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(row)
```

```python
# DictWriter (Advanced):

import csv

with open("students.csv","w", newline="") as file:
    fieldnames = ["name","age","city"]

    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerow({"name":"Rahul","age":21,"city":"Jaipur"})
    writer.writerow({"name":"Amit","age":22,"city":"Delhi"})
```

```python
# Real Example (Filtering Data)
import csv

with open("students.csv","r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        if row["city"] == "Jaipur":
            print(row["name"])
```

```python
# Practice Tasks (Must Do)
# Task 1

# Create CSV file:
# name,marks
# Rahul,85
# Amit,90
# Neha,78

# Print students with marks > 80
# Add one more student using append

import csv
with open("students.csv","w", newline="") as file:
    fieldnames = ["name","marks"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({"name":"Rahul","marks":85})
    writer.writerow({"name":"Amit","marks":90})
    writer.writerow({"name":"Neha","marks":78})

with open("students.csv","r") as file:
  reader = csv.DictReader(file)
  for row in reader:
    if int(row["marks"]) > 80:
      print(row["name"])


with open("students.csv","a",newline="") as file:
  data=csv.writer(file)
  data.writerow(["Sneha",80])

with open("students.csv","r") as file:
  reader = csv.reader(file)
  for row in reader:
    print(row)
```
