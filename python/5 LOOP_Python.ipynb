```python
# Introduction to looping statement :

# a loop is used to repeat a block of code multiple times.
#  The two main types of loops are "FOR" and "WHILE".

```

```python
# 1. FOR Loop :

# Used to iterate over a sequence (list, tuple, string, range, etc.).

for i in range(5):
    print(i)
```

```python
# Nested Loop : A loop inside another loop.

for i in range(1, 4):
    for j in range(1,4):
        print(i, j)
```

```python
# Loop through a list:
fruits = ["apple", "banana", "mango"]

for i in fruits:
    print(i)
```

```python
# else with Loop : Runs when the loop finishes normally (without break).

for i in range(3):
    print(i)
else:
    print("Loop finished")
```

```python
for i in range(1,11,2):
  print(i)
```

```python
# print all even number between 1 to 20
for i in range(1,21):
  if(i%2==0):
    print(i)
```

```python
# print all odd number between 1 to 20
for i in range(1,21):
  if(i%2!=0):
    print(i)
```

```python
# Square of each number from 1 to 20
for  i in range(1,16):
  i=i**2
  print(i, end=" ")
```

```python
# print each character of a string 'PYTHON'
a='PYTHON'
for i in range(len(a)):
  print(a[i])
```

```python
# calculate the product of numbers from 1 to 5
product=1
for i in range(1,6):
  product=product*i
  print(product)
```

```python
# sum of digit of a number

num='1234'
sum=0
for i in range(len(num)):
  sum=sum+int(num[i])
  print(sum)
```

```python
# reverse a string using for loop

a="Hello python"
b=' '
for i in range(len(a)):
  b=a[i]+b
  print(b)
```

```python

# 2. WHILE Loop :

# Runs as long as a condition is True.

# Example:
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
# break statement : (stop the loop):

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue (skip current iteration):

for i in range(5):
    if i == 2:
        continue
    print(i)
```
