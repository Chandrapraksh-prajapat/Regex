```python
# # Exception Handling : An exception is an error that occurs during the execution of a program.

                      # Instead of crashing the program, we handle the error gracefully using exception handling.


# Why Do We Need Exception Handling?

# Prevent program crash
# Improve user experience
# Handle unexpected input
# Maintain program flow
# Professional coding practice
# In real software → exception handling is mandatory.


# Basic Syntax
# try:
#     # risky code
# except SomeError:
#     # handle error



# Types of Common Exceptions

# ZeroDivision:       Error	Divide by 0
# ValueError:	        Wrong value type
# TypeError:    	    Wrong data type
# IndexError:   	    Invalid list index
# KeyError:	          Wrong dictionary key
# FileNotFoundError:	File missing




# blocks:

# try block
# except block
# else block : will be executed when no error in program
# finally block: always execute

#raise keywords is use to create a manual error
```

```python
# Without Exception Handling

a = 10
b = 0

print(a / b)

# Output:

# ZeroDivisionError: division by zero

# Program stops immediately
```

```python
# With Exception Handling
a = 10
b = 0

try:
    print(a / b)
except ZeroDivisionError:
    print("Cannot divide by zero")

# Output:

# Cannot divide by zero
```

```python
#nested try except
try:
  a=100
  try:
    b=10
    print(a/b)
  except:
    print("Inner error handling")

  a+"asssw"
  print("Output is here")
  # b=c/d
except Exception as a:
  print("Error occured:",a)
```

```python
# Multiple Exceptions
try:
    num = int(input("Enter number: "))
    print(10 / num)

except ZeroDivisionError:
    print("Zero not allowed")

except ValueError:
    print("Invalid input")
```

```python
# Using as Keyword
try:
    a = 10 / 0
except ZeroDivisionError as e:
   print("Error occurred:", e)
```

```python
# Using finally

# finally block always executes.

try:
    print(10 / 0)
except:
    print("Error handled")
finally:
    print("Execution finished")


# Output:
# Error handled
# Execution finished

# Even if error happens → finally runs.

# Used for:
# Closing files
# Closing database connections
```

```python
#  else Block

# else runs when NO exception occurs.

try:
    num = int(input("Enter number: "))
except ValueError:
    print("Invalid input")
else:;/.j
    print("You entered:", num)
```

```python
# Important Example (Realistic)
try:
    age = int(input("Enter your age: "))

    if age < 0:
        raise ValueError("Age cannot be negative")

    print("Your age is:", age)

except ValueError as e:
    print("Error:", e)
```

```python
# Raising Custom Exception

# We can manually raise errors.

raise Exception("Something went wrong")
```

```python
#raise keyword

account=15000
amount=105000
try:
  if(amount>15000):
    print("account can be open")
  else:
    raise ValueError

except:
  print("Value is less than 15000")
  # print("account can not be open")
```

```python
# Creating Custom Exception Class (Advanced)

class MyError(Exception):
    pass

try:
    raise MyError("Custom error occurred")
except MyError as e:
    print(e)

# This is used in real projects.
```

```python
# Interview Questions

# Q1: Difference between error and exception?

  # Error → Syntax mistake (compile time)

  # Exception → Runtime problem


# Q2: Difference between finally and else?

  # else → Runs if no error

  # finally → Always runs


# Q3: Why use exception handling?

  # To avoid crash and manage runtime errors properly.
```
