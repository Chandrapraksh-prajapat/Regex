```python
# Find LCM
def LCM(x,y):
  if x>y :
    max=y
  else:
    max =x
  for i in range( max , (x*y)+1 ):
    if (i%x ==0 and i%y ==0):
      print(i)
      break
LCM(10,15)
```

```python
# Built-in Higher Order Functions:

# map()
# filter()
# reduce()


# map(): map() applies a function to each element of an iterable (like list) and returns a new result.(#lazy evaluation)
#       saving memory, especially with large datasets.
#         To view the results as a list, tuple, or other collection,
#           you can convert the map object using functions like list(), tuple(), etc.



#syntax:
        # map(function, iterable)




# Map vs for loop :-

                # for loop process the code at the same time
                # map process the code on demand
```

```python
numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)

print(list(result))
```

```python

# Filter() : filter() selects elements from iterable based on condition.


# syntax:
# filter(function, iterable)


# Get even numbers
nums = [1,2,3,4,5,6]

even = list(filter(lambda x: x%2==0, nums))
print(even)
```

```python
# Filter numbers greater than 3

nums = [1,2,3,4,5]

res = list(filter(lambda x: x>3, nums))
print(res)
```

```python
# Lambda:-

# A lambda function is a small anonymous function (function without name) used for short operations.

# It is written in one line.

# Syntax
# lambda arguments : expression


# use :
#  Short functions
#  Used with map/filter
#  Clean code



# Normal Function vs Lambda

def add(a,b):
    return a+b



add = lambda a,b : a+b
print(add(2,3))
```

```python
# # Generator : A Generator is a special type of function that returns values one by one instead of all at once.

# # It uses yield instead of return.


# Why Generators?
# Normal function: stores ALL values in memory
#   Generator:produces values when needed (lazy loading)
#     So generators are memory efficient.


# Jab function me return ki jagah yield likhte ho, to function ek generator ban jata hai.
```

```python
def numbers():
    return [1,2,3]

print(numbers())
```

```python
def numbers():
    yield 1
    yield 2
    yield 3

for n in numbers():
    print(n)
```

```python
def count(n):
    for i in range(n):
        yield i

for x in count(5):
    print(x)
```

```python
gen = (x*x for x in range(5))
print(list(gen))
```

```python
# Why generators are memory efficient?

# Because they generate values on demand instead of storing all values in memory.
```
