```python
# String in Python : In Python, a string is a sequence of characters used to represent text.
                    # Strings are immutable (cannot be changed after creation)
                    # indexing start from 0
                    # length start from 1
                    #  Reverse indexing start from -1
                    # we can perfrom slicing in list.



#Creating strings : You can create strings using
                    # Single quotes,
                    # Double quotes,
                    # Triple quotes

s1 = 'Hello'
s2 = "Hello"
s3 = """Hello, world!"""
print(s1)
print(s2)
print(s3)
```

```python
# Slicing in String :

# In Python, string slicing is used to extract a part (substring) of a string using index positions.

# string[start : end : step]

# start → index to begin (inclusive)

# end → index to stop (exclusive)

# step → jump/skip count (optional)


s = "Python"

print(s[0:4])   # Pyth
print(s[2:6])   # thon
print(s[:3])    # Pyt
print(s[3:])    # hon
```

```python
a="i love python"

print(a[0:6])
print(a[7:])
print(a[:6])
print(a[::])
print(a[::2])
print(a[::-1])
```
